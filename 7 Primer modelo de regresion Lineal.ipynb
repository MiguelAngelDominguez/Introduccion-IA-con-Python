# Mi primer modelo de regresion lineal

La regresion es uno de los primero algoritmos de machine learning y los podemos usar cuando tenemos datos que muestran una relacion que se puede representar por una linea.



Tenemos dos ejes el peso y la altura, donde se muestra un colecta de datos, sin relacion alguna, pero a partir de estos datos podemos empezar a intuir futuros datos; hay una relacion que se establecera, esta relacion para todos los datos sera una linea que media de conformidad y de aproximacion a futuros datos entrantes.

Crearemos un modelo basico de Fahrenheit y de Celsius, a pezar de que la relacion sea muy basica, nos servira para iniciar con nuestro primer modelo.

Para comenzar con la elaboración de nuestro primero modelo, necesitaremos un pequeño conjunto de datos con ejemplos





```python
import pandas as pd
datos = pd.read_csv('./data/celsius.csv')
```

```python
# Con esta linea de codigo podemos ver la 
# informacion de nuestro dataset
datos.info()
```

```python
# Con esta linea de codigo podemos ver los datos en una tabla
datos.head()
```

```python
import seaborn as sb

sb.scatterplot(x="celsius", y="fahrenheit", data = datos
               , hue = "fahrenheit", palette="coolwarm")
```

```python
# Tenemos que separar las caracteristicas(x) de las etiquetas (y)

x = datos["celsius"]
y = datos["fahrenheit"]
```

```python
y
```

```python
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
# Entrenamiento del modelo

modelo.fit(x_procesada,y_procesada)
```

```python
# Ya que el modelo fue entrenado, hacemos una nueva consulta con esta linea de codigo:
new_celsius = int(input("Ingrese el valor en Celsius: "))
prediccion = modelo.predict([[new_celsius]])
print(f"Celcius: {new_celsius} -> Fahrenheit: {prediccion[0][0]}")
```

```python
# Ahora usando la funcion .score() podemos ver el porcentaje de precision de nuestro modelo
modelo.score(x_procesada, y_procesada)
```
